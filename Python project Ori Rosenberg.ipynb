```python
class Node:
    def __init__(self,name,neighbour):
        self.neighbour = {}
        self.name = name
        self.neighbour = dict(neighbour)

    def __str__(self):
        return f"Node {self.name} neighbours and weights {self.neighbour}"

    def __len__(self):
        return len(self.neighbour)

    def __contains__(self, item):
        return item in self.neighbour

    def __getitem__(self, item):
            return self.neighbour[item]

    def __eq__(self, other):
        return self.name == other.name

    def __ne__(self, other):
        return self.name != other.name

    def is_neighbor(self,name):
        if name.name in self.neighbour:
            print(name,'is a neighbour')
        else:
            print(name, 'is not a neighbour')

    def update(self,name,weight):
        if self.name == name:
            print("Node can not be a neighbour of itself",)
        elif self.name != name and name not in self.neighbour:
            self.neighbour.update({name: weight})
        elif self.name != name and name in self.neighbour:
            for key,value in self.neighbour.items():
                if key == name:
                    self.neighbour[key] = max(weight,self.neighbour[key])
                

    def remove_neighbour(self,name):
        if name.name in self.neighbour:
            return self.neighbour.pop(name.name)
            
    def get_neighbour_value(self,name):
        if name.name in self.neighbour:
            return self.neighbour[name.name]

    def is_isolated(self):
        if None in self.neighbour:
            print("This node has no neighbours")
        else:
            print("This node has neighbours")


N1 = Node('N1',{'N7':15,'N6':5,'N5':20,'N4':20,'N2':10})
N2 = Node('N2',{'N4':10,'N3':5})
N3 = Node('N3',{'N4':5,'N2':15})
N4 = Node('N4',{'N5':10})
N5 = Node('N5',{'N6':5})
N6 = Node('N6',{})
N7 = Node('N7',{'N6':10})
N8 = Node('N8',{'N2':20,'N1':5,'N7':5})
N9 = Node('N9',{'N10':10,'N2':15,'N8':20})
N10 = Node('N10',{'N3':15,'N2':5})
```
